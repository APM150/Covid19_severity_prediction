```python
import load_data
import torch
from Project_Models import RNN_Model
import torch.nn as nn
```

```python
df = load_data.load_county_level()
```

```python
print(df[['CountyName', 'StateName', '#Hospitals', '#ICU_beds', 'cases', 'deaths']].head(2).values)
```

```python
print([i.tolist() for i in df['cases'].head(1).values])
```

```python
result = []
for cityInfo in df[['countyFIPS'] + ['#Hospitals', '#ICU_beds']].head(10).values:
    allCityInfo = []  # form the sequence
    for dayICases, dayIDeath in zip(df[(df['countyFIPS']==cityInfo[0])]['cases'].values[0], df[(df['countyFIPS']==cityInfo[0])]['deaths'].values[0]):
        oneCityOneDay = cityInfo[1:].tolist()
        oneCityOneDay.append(dayICases)
        oneCityOneDay.append(dayIDeath)
        allCityInfo.append(oneCityOneDay)
    result.append(allCityInfo)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
result = torch.tensor(result, dtype=torch.float32).to(device)
```

```python
result
```

```python
model = RNN_Model.RNN(4, 128, 2).to(device)
```

```python
criterion = nn.MSELoss()
learning_rate = 0.005
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
print(model(result))
```

```python
print(df['tot_cases'].values)
```

```python
y = []
for totalCases, totalDeaths in zip(df['tot_cases'].head(10).values, df['tot_deaths'].head(10).values):
    y.append([totalCases, totalDeaths])
y = torch.tensor(y, dtype=torch.float32).to(device)
print(y)
```

```python
print(criterion(model(result), y))
```

```python
for epoch in range(10):
    outputs = model(result)
    loss = criterion(outputs, y)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
```
